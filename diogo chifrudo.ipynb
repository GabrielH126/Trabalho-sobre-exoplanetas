<div><span><span style="font-family: Times; font-size: large; font-weight: bold;">Resumo</span><br /></span></div><div style="text-align: center;"><span><div style="font-family: Times; font-size: medium; text-align: justify;">A proposta deste trabalho é fazer os cálculos e a análise de alguns dados planetários, tais quais, o semieixo maior do baricentro do sistema, o semieixo do sistema em si, a massa e o raio do planeta, em função dos de Júpiter e da Terra, probabilidade e tempo de trânsito, grau de insolação, temperatura de equilíbrio, zona de habitabilidade do sistema, interna, central e externa, além da densidade e compara-la com os corpos do sistema solar.</div><div style="font-family: Times; font-size: medium; text-align: justify;"><br /></div><div style="font-family: Times; font-size: medium; text-align: justify;">Todos estes dados serão determinados a partir de planetas descobertos pelo método da velocidade radial, e para tal determinação são necessários os seguintes parâmetros: excentricidade, período orbital, amplitude da velocidade radial, massa, raio e temperatura da estrela, além da profundidade de trânsito. Todos os dados foram retirados de site <a href="https://exoplanetarchive.ipac.caltech.edu/index.html" target="_blank">NASA Exoplanet Archive</a>.</div></span></div><div style="font-family: Times; font-size: medium;"><br /></div><div style="font-family: Times; font-size: medium;">Os sistemas escolhidos foram 55 Cancri, LTT 3780 e K2-285, nos dois primeiros sistemas, existem planetas que são considerados "superterras", planetas rochosos que se assemelham a Terra, e que talvez, possam conter vida, sendo 55 Cancri e, o primeiro exoplaneta desse tipo descoberto, além de todos os sistemas terem os dados necessários para os cálculos que serão feitos. Estes que por sua vez, utilizam metódos conhecidos, e já previamente determinados, mas que serão discutidos a seguir, e todo o trabalho com os valores foram feitos pelo programa a seguir, idealizado com auxílio das bibliotecas Math, Pandas e Scipy.</div>

<div><span><span style="font-family: Times; font-size: large; font-weight: bold;">Fórmulas</span><br /></span></div><div style="text-align: center;"><span style="font-family: Times; font-size: medium;"><div style="text-align: justify;">A discussão das fórmulas utilizadas serão feitas aqui, e seu uso será extensivo, citando sempre esta parte para justificar os resultados encontrados.</div></span></div><b><span style="font-family: Times; font-size: medium;">Semieixo maior</span></b><div style="text-align: justify;"><span style="font-family: Times; font-size: medium;">A partir da terceira lei de Kepler</span></div><div style="text-align: justify;"><span style="font-family: Times; font-size: medium;"><br /></span></div><div style="text-align: justify;"><span style="font-family: Times; font-size: medium;">$\Large a^3 = \Large\frac{P^2}{K}$ </span></div><div style="text-align: justify;"><span style="font-family: Times; font-size: medium;"><br /></span></div><div style="text-align: justify;"><span style="font-family: Times; font-size: medium;">Donde<br /><br /></span></div><div style="text-align: justify;"><span style="font-family: Times; font-size: medium;">$\Large a = \Large(\frac {GM_{S}P^2} {4π^2} )^ \frac {1} {3}$ </span></div><div><span style="font-family: Times; font-size: medium;"><br /></span></div><div><span style="font-family: Times; font-size: medium;">O resultado é dado em metros, o período usado é em dias, e a massa da estrela dada em quilos.</span></div>


```python
import scipy.constants as sci
import pandas as pd
import math


raio_sol = 6.96e8
lumin_sol = 3.84e26
massa_sol = 1.99e30
UA = 1.49e11
massa_terra = 5.9e24
massa_jupiter = 1.9e27
raio_jupiter = 6.99e7
raio_terra = 6.3e6

def semieixo(massa_estrela, periodo):
    sma = (massa_estrela*periodo**2)**(1/3)
    return sma
```

<b style="font-family: Times; font-size: large;">Densidade</b><div><span style="font-family: Times; font-size: medium;">Tendo em vista que a densidade é dada por</span></div><div style="text-align: justify;"><span><span style="font-family: Times; font-size: medium;"><br /></span></span></div><div style="text-align: justify;"><span><span style="font-family: Times; font-size: medium;">$\Large ρ = \frac{M_{P}}{V}$</span></span></div><div style="text-align: justify;"><span style="font-family: Times; font-size: medium;"><br /></span></div><div style="text-align: justify;"><span style="font-family: Times; font-size: medium;">E considerando o planeta como uma esfera, temos que</span></div><div style="text-align: justify;"><span style="font-family: Times; font-size: medium;"><br /></span></div><div style="text-align: justify;"><span style="font-family: Times; font-size: medium;">$\Large ρ = \frac{M_{P}}{\frac {4}{3}πR_{P}^3}$</span></div>

```python
def densidade(massa_planeta, raio_planeta):
    dens = massa_planeta/((4/3)*sci.pi*raio_planeta**3)
    return dens
```

<b><span style="font-family: Times; font-size: large;">Insolação</span><br /></b><span style="font-family: Times; font-size: medium;">É o grau de insolação que um planeta recebe de sua estrela mãe. A quantidade de energia que atravessa uma unidade de área por tempo. Sabendo que o fluxo de radiação emitido por uma estrela é<br /><br /></span><div><span style="font-family: Times; font-size: medium;">$\Large\int_{0}^{∞}Fv dv = S_{P} $<br style="font-family: Times;" /><br /></span></div><div><span style="font-family: Times; font-size: medium;">Logo<br /><br /></span></div><div><span style="font-family: Times; font-size: medium;">$\Large S_{P} = \frac {L_{S}}{4πR^2}$<br style="font-family: Times;" /><br /></span></div><div><span style="font-family: Times; font-size: medium;">E dividindo pelos dados solares, encontra-se uma relação da insolação, em função da insolação terrestre.<br /><br /></span></div><div><span style="font-family: Times; font-size: medium;">$\Large\frac {S_{P}}{S_{⨁}} = \frac {L_{S}{(UA)^2}}{L_{☉}a^2}$</span></div>

```python
def insolacao(lumin, semieixo):
    ins = (lumin/lumin_sol)*(1/(semieixo**2))
    return ins
```

<div style="text-align: justify;"><b><span style="font-family: Times; font-size: large;">Temperatura de equilíbrio&nbsp;</span><br /><span style="font-family: Times; font-size: medium; font-weight: 400;"> A temperatura que o planeta teria, considerando apenas a energia que ele recebe da estrela, e não levando em consideração outros fatores como atmoesfera. Usando com base a ideia anterior para calcular a insolação e considerando um planeta como um corpo negro, supondo que toda a energia absorvida, é igual a energia emitida, então a absorvida é</span></b></div><div style="font-family: Times; font-size: medium; text-align: justify;"><b><span style="font-weight: 400;"><br /></span></b></div><div style="font-family: Times; font-size: medium; text-align: justify;"><b><span style="font-weight: 400;">$\Large P_{i n} = \Large S_{P}(1-A)πRp^2$</span></b></div><div style="text-align: justify;"><b><span style="font-family: Times; font-size: medium; font-weight: 400;"><br /></span></b></div><div style="text-align: justify;"><b><span style="font-family: Times; font-size: medium; font-weight: 400;">E a que emitida</span></b></div><div style="text-align: justify;"><b><span style="font-weight: 400;"><br /></span></b></div><div style="text-align: justify;"><b><span style="font-weight: 400;">$ \Large P_{out} = 4πR_{P}^2 σT_{eq}^4$ </span></b></div><div style="text-align: justify;"><b><span style="font-family: Times; font-size: medium; font-weight: 400;"><br /></span></b></div><div style="text-align: justify;"><b><span style="font-family: Times; font-size: medium; font-weight: 400;">Igualando as duas e isolando a temperatura de equilíbrio, tem-se</span></b></div><div style="text-align: justify;"><b><span style="font-weight: 400;"><br /></span></b></div><div style="text-align: justify;"><b><span style="font-weight: 400;">$\Large T_{eq} = T_{es}(1-A)^ \frac {1} {4} ( \frac {R_{S}} {2a} )^\frac{1}{2}$</span></b></div>

```python
def temperatura_eq(temperatura, raio_estrela, semieixo):
    temperatura_eq = temperatura*((0.633)**(1/4))*((raio_estrela*raio_sol)/(2*semieixo*UA))**(1/2)
    return temperatura_eq
```

<div style="text-align: justify;"><b><span style="font-family: Times; font-size: large;">Zona habitável</span></b></div><div style="text-align: justify;"><b><span style="font-family: Times; font-size: medium; font-weight: 400;">A definição de zona habitável é a área em torno de uma estrela onde a água pode existir no estado líquido, ou seja, a temperatura do planeta deve estar entre 273K e 373K, utilizando a fórmula da luminosidade&nbsp;<br /><br /></span></b></div><div style="text-align: justify;"><b><span style="font-family: Times; font-size: medium; font-weight: 400;">$ \Large L_{S} = 4πR_{P}^2 σT_{ef}^4$<div style="font-family: Times; font-size: medium;"><br /></div><div style="font-family: Times; font-size: medium;">Logo, pode-se determinar a zona de habitabilidade interna, central, e externa, pelas seguintes relações<br />$\Large Hz_{int} = 0,75(\frac {L_{s}}{L_{☉}})^ \frac {1}{2}$</div><br />$\Large Hz_{mid} = (\frac {L_{s}}{L_{☉}})^ \frac {1}{2}$</span></b></div><br />$\Large Hz_{ext} = 1,77(\frac {L_{s}}{L_{☉}})^ \frac {1}{2}$

```python
def luminosidade(raio_estrela, temperatura):
    lumin = (4*sci.pi)*((raio_estrela)**2)*(sci.sigma)*(temperatura**4)
    return lumin

def habitabilidade(lumin):
    H_interno = 0.75*(lumin/lumin_sol)**(1/2)
    H_central = (lumin/lumin_sol)**(1/2)
    H_externo = 1.77*(lumin/lumin_sol)**(1/2)
    return H_interno, H_central, H_externo
```

<div style="text-align: justify;"><b><span style="font-family: Times; font-size: large;">Raio do planeta</span></b></div><div style="text-align: justify;"><b><span style="font-family: Times; font-size: medium; font-weight: 400;">Sabendo que o fluxo estelar diminui com um fator de</span></b></div><div style="text-align: justify;"><b><span style="font-family: Times; font-size: medium; font-weight: 400;"><br /></span></b></div><div style="text-align: justify;"><b><span style="font-family: Times; font-size: medium; font-weight: 400;">$\Large L_{P} = L_{S} - L_{P'}$<br /><br /></span></b></div><div style="text-align: justify;"><b><span style="font-family: Times; font-size: medium; font-weight: 400;">$\Large L_{P'} = 4πR_{P}^2 F_{S}$</span></b></div><div style="text-align: justify;"><b><span style="font-family: Times; font-size: medium; font-weight: 400;"><br /></span></b></div><div style="text-align: justify;"><b><span style="font-family: Times; font-size: medium; font-weight: 400;">A variação desse fluxo pode ser escrita por</span></b></div><div style="text-align: justify;"><b><span style="font-family: Times; font-size: medium; font-weight: 400;"><br /></span></b></div><div style="text-align: justify;"><b><span style="font-family: Times; font-size: medium; font-weight: 400;">$\Large ΔF = \frac{4πR_{P}^2 F_{S}}{4πR_{S}^2 F_{S}}$</span></b></div><div style="text-align: justify;"><span style="font-family: Times; font-size: medium;"><br /></span></div><div style="text-align: justify;"><span style="font-family: Times; font-size: medium;">Reduzindo a equação e escrevendo em função do raio de júpiter, chega-se a fórmula de</span></div><div style="text-align: justify;"><span style="font-family: Times; font-size: medium;"><br /></span></div><div style="text-align: justify;"><span style="font-family: Times; font-size: medium;">$\Large T_{depth} = 1,049\frac{R_{P}^2}{R_{S}^2}$</span></div><div style="text-align: justify;"><span style="font-family: Times; font-size: medium;"><br /></span></div><div style="text-align: justify;"><span style="font-family: Times; font-size: medium;">E isolando o raio do planeta, o resultado final é</span></div><div style="text-align: justify;"><span style="font-family: Times; font-size: medium;"><br /></span></div><div style="text-align: justify;"><span style="font-family: Times; font-size: medium;">$\Large R_{P} = R_{S}\sqrt\frac{T_{depth}}{1,049}$</span></div><div style="text-align: justify;"><br /></div>

```python
def raio(trans_depth, raio_estrela):
    raio_planeta = ((trans_depth*raio_estrela**2)/1.049)**(1/2)
    return raio_planeta
```

<div style="text-align: justify;"><b><span style="font-family: Times; font-size: large;">Probabilidade de trânsito</span></b></div><div style="text-align: justify;"><span style="font-family: Times; font-size: medium;">Usando as coordenadas polares, x pode ser escrito da seguinte forma</span></div><div style="text-align: justify;"><span style="font-family: Times; font-size: medium;"><br /></span></div><div style="text-align: justify;"><span style="font-family: Times; font-size: medium;">$\Large x = rcos(i)$<br /><br /></span></div><div style="text-align: justify;"><span style="font-family: Times; font-size: medium;">Onde r = a. A condição necessária para que o trânsito ocorra é<br /><br /></span></div><div style="text-align: justify;"><span style="font-family: Times; font-size: medium;">$\Large acos(i) < R_{S} + R_{P}$<br /><br /></span></div><div style="text-align: justify;"><span style="font-family: Times; font-size: medium;">A partir dessa ideia, o trânsito pode ocorrer até o limite do tamanho estelar, então, a probabilidade que ele ocorra é definida por&nbsp;<br /><br /></span></div><div style="text-align: justify;"><span style="font-family: Times; font-size: medium;">$\Large P_{tr} = \frac {\int_{0}^{máx cos(i)}d(cos(i))}{\int_{0}^{1}d(cos(i))}$<br /></span></div><div style="text-align: justify;"><span style="font-family: Times; font-size: medium;"><br /></span></div><div style="text-align: justify;"><span style="font-family: Times; font-size: medium;">Como o máximo valor que cos(i) pode assumir é</span></div><div style="text-align: justify;"><span style="font-family: Times; font-size: medium;"><br /></span></div><div style="text-align: justify;"><span style="font-family: Times; font-size: medium;">$\Large cos(i) = \frac {R_{S}+R_{P}}{a}$</span></div><div style="text-align: justify;"><span style="font-family: Times; font-size: medium;"><br /></span></div><div style="text-align: justify;"><span style="font-family: Times; font-size: medium;">Tem- se que</span></div><div style="text-align: justify;"><span style="font-family: Times; font-size: medium;"><br /></span></div><div style="text-align: justify;"><span style="font-family: Times; font-size: medium;">$\Large P_{tr} = \frac {R_{S}+R_{P}}{a}\frac{\int_{0}^{1}d(cos(i))}{\int_{0}^{1}d(cos(i))}$</span></div><div style="text-align: justify;"><span style="font-family: Times; font-size: medium;"><br /></span></div><div style="text-align: justify;"><span style="font-family: Times; font-size: medium;">Sabendo que a função cos tem seu máximo em 1, a equação anterior pode ser escrita da seguinte forma</span></div><div style="text-align: justify;"><span style="font-family: Times; font-size: medium;"><br /></span></div><div style="text-align: justify;"><span style="font-family: Times; font-size: medium;">$\Large P_{tr} = \frac {R_{S}+R_{P}}{a}$</span></div><div style="text-align: justify;"><span style="font-family: Times; font-size: medium;"><br /></span></div><div style="text-align: justify;"><span style="font-family: Times; font-size: medium;">E assumindo que Rs &gt;&gt; Rp</span></div><div style="text-align: justify;"><span style="font-family: Times; font-size: medium;"><br /></span></div><div style="text-align: justify;"><span style="font-family: Times; font-size: medium;">$\Large P_{tr} = \frac {R_{S}}{a}$</span></div><div style="text-align: justify;"><span style="font-family: Times; font-size: medium;"><br /></span></div><div style="text-align: justify;"><span style="font-family: Times; font-size: medium;">Mas para esse resultado, não é levada em consideração a excentricidade, pode-se considerar que</span></div><div style="text-align: justify;"><span style="font-family: Times; font-size: medium;"><br /></span></div><div style="text-align: justify;"><span style="font-family: Times; font-size: medium;">$\Large r = \frac{a(1-e^{2})}{1+ecos(θ)}$</span></div><div style="text-align: justify;"><span style="font-family: Times; font-size: medium;"><br /></span></div><div style="text-align: justify;"><span style="font-family: Times; font-size: medium;">Então</span></div><div style="text-align: justify;"><span style="font-family: Times; font-size: medium;"><br /></span></div><div style="text-align: justify;"><span style="font-family: Times; font-size: medium;">$\Large P_{tr} = \frac {R_{S}+R_{P}(1+ecos(θ))}{a(1-e^{2})}$</span></div><div style="text-align: justify;"><span style="font-family: Times; font-size: medium;"><br /></span></div><div style="text-align: justify;"><span style="font-family: Times; font-size: medium;">E considerando que o cos(<b style="background-color: white; text-align: left;">θ</b>) é aproximadamente igual a 0, a equação para a probabilidade de trânsito é</span></div><div style="text-align: justify;"><span style="font-family: Times; font-size: medium;"><br /></span></div><div style="text-align: justify;"><span style="font-family: Times; font-size: medium;">$\Large P_{tr} = \frac {R_{S}+R_{P}}{a(1-e^{2})}$</span></div>

```python
def probabilidade_transito(raio_estrela, semieixo, excentricidade):
    prob = (raio_estrela*raio_sol)/(semieixo*UA*(1-excentricidade)**2)
    return prob*100
```

<span style="font-family: Times; font-size: large;"><b>Massa do planeta<br /></b></span><div><span style="font-family: Times; font-size: medium;">Imaginando que o planeta e a estrela possuem o mesmo período orbital, logo, eles teriam o mesmo momento, então, o seguinte é válido</span></div><div><span style="font-family: Times; font-size: medium;"><br /></span></div><div><span style="font-family: Times; font-size: medium;">$\Large M_{S}v_{S} = M_{P}v_{P}$</span></div><div><span style="font-family: Times; font-size: medium;"><br /></span></div><div><span style="font-family: Times; font-size: medium;">Determinando a massa e a velocidade da estrela, resta apenas trabalhar na velocidade do planeta, para determinar sua massa, e utilizando a equação anterior, para determinar o semieixo maior, tal que</span></div><div><span style="font-family: Times; font-size: medium;"><br /></span></div><div><span style="font-family: Times; font-size: medium;">$\Large v_{P} = \frac {2πa}{P}$</span></div><div><span style="font-family: Times; font-size: medium;"><br /></span></div><div><span style="font-family: Times; font-size: medium;">Acima, a velocidade do planeta, é dada pela distância que ele percorre, circunferência da orbita, sobre o tempo em que ele faz esse caminho, o período, que é observável, substituindo a velocidade, na primeira equação, e usando a velocidade da estrela como velocidade radial, é possível encontrar que</span></div><div><span style="font-family: Times; font-size: medium;"><br /></span></div><div><span style="font-family: Times; font-size: medium;">$\Large M_{P}= \frac {M_{S}KP}{2πa}$</span></div>

```python
def massa(massa_estrela, amp_velocidade, periodo, semieixo):
    massa_planeta = (massa_estrela*amp_velocidade*periodo)/(2*sci.pi*semieixo)
    return massa_planeta
```

<span style="font-family: Times;"><b style="font-size: x-large;">Tempo de trânsito<br /></b><span style="font-family: Times; font-size: medium;">Tempo que o planeta leva para atravessar completamente sua estrela.</span></span><div><span style="font-family: Times; font-size: medium;"><div class="separator" style="clear: both; text-align: center;"><div class="separator" style="clear: both; text-align: left;"><a href="https://i.imgur.com/c0RWXYD.png" style="margin-left: 1em; margin-right: 1em;"><img border="0" data-original-height="237" data-original-width="268" height="237" src="https://i.imgur.com/c0RWXYD.png" width="268" /></a></div></div></span></div><div><span style="font-family: Times;"><span style="font-family: Times; font-size: medium;">A partir do desenho acima, tem-se o corpo maior como a estrela, e o menor como o planeta, o raio da estrela mais o raio do planeta, compõem a hipotenusa desse triângulo, acos(i), ou x, faz parte de um cateto, e no outro, aparece uma variável I, que é metade do caminho que o planeta irá percorrer na frente da estrela, sendo 2I, todo o trajeto, por Pitágoras</span></span></div><div><span style="font-family: Times;"><span style="font-family: Times; font-size: medium;"><br /></span></span></div><div><span style="font-family: Times;"><span style="font-family: Times; font-size: medium;">$\Large ((R_{P}+R_{S})^2 - (acos(i))^{2})^{1/2} = I$</span></span></div><div><span style="font-size: medium;"><span style="font-family: Times;"><span style="font-family: Times;"><br /></span></span></span></div><div><span style="font-size: medium;"><span style="font-family: Times;"><span style="font-family: Times;">O tempo de trânsito depende em qual porção da estrela o planeta atravessa, sendo a porção equatorial, o tempo máximo, onde&nbsp;</span></span><span style="font-family: Times;">acos(i) = 0, nas outras porções, o tempo será menor que esse.</span></span></div><div><span style="font-size: medium;"><span style="font-family: Times;">Considerando uma visão lateral, o planeta parecerá mover-se de um ponto A para um ponto B, fazendo um ângulo&nbsp;</span><span style="background-color: white;"><span style="font-family: Times;">α com o centro da estrela. Então, por trigonometria, tem-se</span></span></span></div><div><span style="font-size: medium;"><span style="background-color: white;"><span style="font-family: Times;"><br /></span></span></span></div><div><span style="background-color: white;"><span style="font-family: Times; font-size: medium;">$\Large sen\frac{α}{2} = \frac {I}{sen(i)}$</span></span></div><div><span style="background-color: white;"><span style="font-family: Times; font-size: medium;"><br /></span></span></div><div><span style="background-color: white;"><span style="font-family: Times; font-size: medium;">Substituindo 1 pela equação anterior</span></span></div><div><span style="background-color: white;"><span style="font-family: Times; font-size: medium;"><br /></span></span></div><div><span style="background-color: white;"><span style="font-family: Times; font-size: medium;">$\Large sen\frac{α}{2} = \frac {((R_{P}+R_{S})^2 - (acos(i))^{2})^{1/2}}{asen(i)}$</span></span></div><div><span style="background-color: white;"><span style="font-family: Times; font-size: medium;"><br /></span></span></div><div><span style="background-color: white;"><span style="font-family: Times; font-size: medium;">E reorganizando a equação, obtem-se</span></span></div><div><span style="background-color: white;"><span style="font-family: Times; font-size: medium;"><br /></span></span></div><div><span style="background-color: white;"><span style="font-family: Times; font-size: medium;">$\Large sen\frac{α}{2} = \frac {R_{S}}{a}\frac {((1+ \frac {R_{P}}{R_{S}})^2 - (\frac {acos(i)}{R_{S}})^{2})^{1/2}}{1-cos(i)^2}$</span></span></div><div><span style="font-size: medium;"><span style="background-color: white;"><span style="font-family: Times;"><br /></span></span></span></div><div><span style="font-size: medium;"><span style="background-color: white;"><span style="font-family: Times;">Como o&nbsp;</span></span><span style="font-family: Times;">ângulo&nbsp;</span><span style="background-color: white;"><span style="font-family: Times;">α é relacionado com o comprimento da circunferência pelo período, onde</span></span></span></div><div><span style="font-size: medium;"><span style="background-color: white;"><span style="font-family: Times;"><br /></span></span></span></div><div><span style="background-color: white;"><span style="font-family: Times; font-size: medium;">$\Large α = \frac {2πt_{r}}{P}$</span></span></div><div><span style="background-color: white;"><span style="font-family: Times; font-size: medium;"><br /></span></span></div><div><span style="background-color: white;"><span style="font-family: Times; font-size: medium;">Então</span></span></div><div><span style="background-color: white;"><span style="font-family: Times; font-size: medium;">$\Large sen\frac {πt_{r}}{P}$</span></span></div><div><span style="background-color: white;"><span style="font-family: Times; font-size: medium;"><br /></span></span></div><div><span style="background-color: white;"><span style="font-family: Times; font-size: medium;">E por consequência, isolando Tr é dado por</span></span></div><div><span style="background-color: white;"><span style="font-family: Times; font-size: medium;"><br /></span></span></div><div><span style="background-color: white;"><span style="font-family: Times; font-size: medium;">$\Large T_{t} = \frac{P}{π}arcsen(\frac{R_{S}}{a}[\frac{(1-\frac {R_{P}}{R_{S}})^{2} - (\frac{a}{R_{S}}cos(i)^{2})^2}{1-cos(i)^{2}}]^{1/2})$</span></span></div>

```python
def tempo_transito(periodo, raio_estrela, semieixo, raio_planeta, inclinacao):
    parte1 = ((semieixo/raio_estrela)*(math.cos(math.radians(inclinacao))**2))**2
    parte2 = 1-(math.cos(math.radians(inclinacao))**2)
    parte3 = (1-raio_planeta/raio_estrela)**2
    final = (periodo/math.pi)*math.asin((raio_estrela/semieixo)*(((parte3-parte1)/parte2)**(1/2)))
    return final
```

<p style="text-align: justify;"><span style="font-family: Times;"><span style="font-size: large;"><b>Semieixo maior do baricentro do sistema<br /></b></span><span style="font-size: medium;">Um baricentro, ou centroide, para a geometria, é o ponto em que todas as medianas de um polígono&nbsp;se encontram, já do ponto de vista astrofísico, ele seria o centro da gravidade, ou mais conhecido como centro de massa, ponto onde as massas estão em equilíbrio, num sistema planetário, é o local em torno do qual os corpos do sistema orbitam. Fazendo a aproximação das órbitas como esférica, usa-se essa aproximação para ser determinado.</span></span></p><p style="text-align: justify;"><span style="font-family: Times;"><span style="font-size: medium;">Fazendo certas considerações, buscando facilitar o problema, e usando coordenadas cartesianas a posição da estrela pode ser dada como (0,0), a posição do exoplaneta como (x, 0). E utilizando as noções básicas do centro de massa, chegasse a seguinte relação para encontrar o semieixo do baricentro.</span></span></p><p style="text-align: justify;"><span><span style="font-size: medium;">$\large B = \frac {P(1-e^{2})^{\frac{1}{2}}K}{2πsen(i)}$</span><br /><br /></span></p>

```python
def baricentro(periodo, excentricidade, velocidade_radial, inclinacao):
    bar_sma = (periodo*((1-(excentricidade**2))**(1/2))*velocidade_radial)/(2*sci.pi*math.sin(math.radians(inclinacao)))
    return bar_sma
```
