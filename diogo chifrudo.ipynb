Todos os dados foram retirados de site <a href="https://exoplanetarchive.ipac.caltech.edu/index.html" target="_blank">NASA Exoplanet Archive</a>.</div></span></div><div style="font-family: Times; font-size: medium;"><br /></div><div style="font-family: Times; font-size: medium;">Os sistemas escolhidos foram 55 Cancri, LTT 3780 e K2-285, nos dois primeiros sistemas, existem planetas que so considerados "superterras", planetas rochosos que se assemelham a Terra, e que talvez, possam conter vida, sendo 55 Cancri e, o primeiro exoplaneta desse tipo descoberto, alm de todos os sistemas terem os dados necessrios para os clculos que sero feitos. Estes que por sua vez, utilizam metdos conhecidos, e j previamente determinados, mas que sero discutidos a seguir, e todo o trabalho com os valores foram feitos pelo programa a seguir, idealizado com auxlio das bibliotecas Math, Pandas e Scipy.</div>

<div><span><span style="font-family: Times; font-size: large; font-weight: bold;">Frmulas</span><br /></span></div><div style="text-align: center;"><span style="font-family: Times; font-size: medium;"><div style="text-align: justify;">A discusso das frmulas utilizadas sero feitas aqui, e seu uso ser extensivo, citando sempre esta parte para justificar os resultados encontrados.</div></span></div><b><span style="font-family: Times; font-size: medium;">Semieixo maior</span></b><div style="text-align: justify;"><span style="font-family: Times; font-size: medium;">A partir da terceira lei de Kepler</span></div><div style="text-align: justify;"><span style="font-family: Times; font-size: medium;"><br /></span></div><div style="text-align: justify;"><span style="font-family: Times; font-size: medium;">$\Large a^3 = \Large\frac{P^2}{K}$ </span></div><div style="text-align: justify;"><span style="font-family: Times; font-size: medium;"><br /></span></div><div style="text-align: justify;"><span style="font-family: Times; font-size: medium;">Donde<br /><br /></span></div><div style="text-align: justify;"><span style="font-family: Times; font-size: medium;">$\Large a = \Large(\frac {GM_{S}P^2} {4^2} )^ \frac {1} {3}$ </span></div><div><span style="font-family: Times; font-size: medium;"><br /></span></div><div><span style="font-family: Times; font-size: medium;">O resultado  dado em metros, o perodo usado  em dias, e a massa da estrela dada em quilos.</span></div>


```python
import scipy.constants as sci
import pandas as pd
import math


raio_sol = 6.96e8
lumin_sol = 3.84e26
massa_sol = 1.99e30
UA = 1.49e11
massa_terra = 5.9e24
massa_jupiter = 1.9e27
raio_jupiter = 6.99e7
raio_terra = 6.3e6

def semieixo(massa_estrela, periodo):
    sma = (massa_estrela*periodo**2)**(1/3)
    return sma
```

<b style="font-family: Times; font-size: large;">Densidade</b><div><span style="font-family: Times; font-size: medium;">Tendo em vista que a densidade  dada por</span></div><div style="text-align: justify;"><span><span style="font-family: Times; font-size: medium;"><br /></span></span></div><div style="text-align: justify;"><span><span style="font-family: Times; font-size: medium;">$\Large  = \frac{M_{P}}{V}$</span></span></div><div style="text-align: justify;"><span style="font-family: Times; font-size: medium;"><br /></span></div><div style="text-align: justify;"><span style="font-family: Times; font-size: medium;">E considerando o planeta como uma esfera, temos que</span></div><div style="text-align: justify;"><span style="font-family: Times; font-size: medium;"><br /></span></div><div style="text-align: justify;"><span style="font-family: Times; font-size: medium;">$\Large  = \frac{M_{P}}{\frac {4}{3}R_{P}^3}$</span></div>

```python
def densidade(massa_planeta, raio_planeta):
    dens = massa_planeta/((4/3)*sci.pi*raio_planeta**3)
    return dens
```

<b><span style="font-family: Times; font-size: large;">Insolao</span><br /></b><span style="font-family: Times; font-size: medium;"> o grau de insolao que um planeta recebe de sua estrela me. A quantidade de energia que atravessa uma unidade de rea por tempo. Sabendo que o fluxo de radiao emitido por uma estrela <br /><br /></span><div><span style="font-family: Times; font-size: medium;">$\Large\int_{0}^{}Fv dv = S_{P} $<br style="font-family: Times;" /><br /></span></div><div><span style="font-family: Times; font-size: medium;">Logo<br /><br /></span></div><div><span style="font-family: Times; font-size: medium;">$\Large S_{P} = \frac {L_{S}}{4R^2}$<br style="font-family: Times;" /><br /></span></div><div><span style="font-family: Times; font-size: medium;">E dividindo pelos dados solares, encontra-se uma relao da insolao, em funo da insolao terrestre.<br /><br /></span></div><div><span style="font-family: Times; font-size: medium;">$\Large\frac {S_{P}}{S_{}} = \frac {L_{S}{(UA)^2}}{L_{}a^2}$</span></div>

```python
def insolacao(lumin, semieixo):
    ins = (lumin/lumin_sol)*(1/(semieixo**2))
    return ins
```

<div style="text-align: justify;"><b><span style="font-family: Times; font-size: large;">Temperatura de equilbrio&nbsp;</span><br /><span style="font-family: Times; font-size: medium; font-weight: 400;"> A temperatura que o planeta teria, considerando apenas a energia que ele recebe da estrela, e no levando em considerao outros fatores como atmoesfera. Usando com base a ideia anterior para calcular a insolao e considerando um planeta como um corpo negro, supondo que toda a energia absorvida,  igual a energia emitida, ento a absorvida </span></b></div><div style="font-family: Times; font-size: medium; text-align: justify;"><b><span style="font-weight: 400;"><br /></span></b></div><div style="font-family: Times; font-size: medium; text-align: justify;"><b><span style="font-weight: 400;">$\Large P_{i n} = \Large S_{P}(1-A)Rp^2$</span></b></div><div style="text-align: justify;"><b><span style="font-family: Times; font-size: medium; font-weight: 400;"><br /></span></b></div><div style="text-align: justify;"><b><span style="font-family: Times; font-size: medium; font-weight: 400;">E a que emitida</span></b></div><div style="text-align: justify;"><b><span style="font-weight: 400;"><br /></span></b></div><div style="text-align: justify;"><b><span style="font-weight: 400;">$ \Large P_{out} = 4R_{P}^2 T_{eq}^4$ </span></b></div><div style="text-align: justify;"><b><span style="font-family: Times; font-size: medium; font-weight: 400;"><br /></span></b></div><div style="text-align: justify;"><b><span style="font-family: Times; font-size: medium; font-weight: 400;">Igualando as duas e isolando a temperatura de equilbrio, tem-se</span></b></div><div style="text-align: justify;"><b><span style="font-weight: 400;"><br /></span></b></div><div style="text-align: justify;"><b><span style="font-weight: 400;">$\Large T_{eq} = T_{es}(1-A)^ \frac {1} {4} ( \frac {R_{S}} {2a} )^\frac{1}{2}$</span></b></div>

```python
def temperatura_eq(temperatura, raio_estrela, semieixo):
    temperatura_eq = temperatura*((0.633)**(1/4))*((raio_estrela*raio_sol)/(2*semieixo*UA))**(1/2)
    return temperatura_eq
```

<div style="text-align: justify;"><b><span style="font-family: Times; font-size: large;">Zona habitvel</span></b></div><div style="text-align: justify;"><b><span style="font-family: Times; font-size: medium; font-weight: 400;">A definio de zona habitvel  a rea em torno de uma estrela onde a gua pode existir no estado lquido, ou seja, a temperatura do planeta deve estar entre 273K e 373K, utilizando a frmula da luminosidade&nbsp;<br /><br /></span></b></div><div style="text-align: justify;"><b><span style="font-family: Times; font-size: medium; font-weight: 400;">$ \Large L_{S} = 4R_{P}^2 T_{ef}^4$<div style="font-family: Times; font-size: medium;"><br /></div><div style="font-family: Times; font-size: medium;">Logo, pode-se determinar a zona de habitabilidade interna, central, e externa, pelas seguintes relaes<br />$\Large Hz_{int} = 0,75(\frac {L_{s}}{L_{}})^ \frac {1}{2}$</div><br />$\Large Hz_{mid} = (\frac {L_{s}}{L_{}})^ \frac {1}{2}$</span></b></div><br />$\Large Hz_{ext} = 1,77(\frac {L_{s}}{L_{}})^ \frac {1}{2}$

```python
def luminosidade(raio_estrela, temperatura):
    lumin = (4*sci.pi)*((raio_estrela)**2)*(sci.sigma)*(temperatura**4)
    return lumin

def habitabilidade(lumin):
    H_interno = 0.75*(lumin/lumin_sol)**(1/2)
    H_central = (lumin/lumin_sol)**(1/2)
    H_externo = 1.77*(lumin/lumin_sol)**(1/2)
    return H_interno, H_central, H_externo
```

<div style="text-align: justify;"><b><span style="font-family: Times; font-size: large;">Raio do planeta</span></b></div><div style="text-align: justify;"><b><span style="font-family: Times; font-size: medium; font-weight: 400;">Sabendo que o fluxo estelar diminui com um fator de</span></b></div><div style="text-align: justify;"><b><span style="font-family: Times; font-size: medium; font-weight: 400;"><br /></span></b></div><div style="text-align: justify;"><b><span style="font-family: Times; font-size: medium; font-weight: 400;">$\Large L_{P} = L_{S} - L_{P'}$<br /><br /></span></b></div><div style="text-align: justify;"><b><span style="font-family: Times; font-size: medium; font-weight: 400;">$\Large L_{P'} = 4R_{P}^2 F_{S}$</span></b></div><div style="text-align: justify;"><b><span style="font-family: Times; font-size: medium; font-weight: 400;"><br /></span></b></div><div style="text-align: justify;"><b><span style="font-family: Times; font-size: medium; font-weight: 400;">A variao desse fluxo pode ser escrita por</span></b></div><div style="text-align: justify;"><b><span style="font-family: Times; font-size: medium; font-weight: 400;"><br /></span></b></div><div style="text-align: justify;"><b><span style="font-family: Times; font-size: medium; font-weight: 400;">$\Large F = \frac{4R_{P}^2 F_{S}}{4R_{S}^2 F_{S}}$</span></b></div><div style="text-align: justify;"><span style="font-family: Times; font-size: medium;"><br /></span></div><div style="text-align: justify;"><span style="font-family: Times; font-size: medium;">Reduzindo a equao e escrevendo em funo do raio de jpiter, chega-se a frmula de</span></div><div style="text-align: justify;"><span style="font-family: Times; font-size: medium;"><br /></span></div><div style="text-align: justify;"><span style="font-family: Times; font-size: medium;">$\Large T_{depth} = 1,049\frac{R_{P}^2}{R_{S}^2}$</span></div><div style="text-align: justify;"><span style="font-family: Times; font-size: medium;"><br /></span></div><div style="text-align: justify;"><span style="font-family: Times; font-size: medium;">E isolando o raio do planeta, o resultado final </span></div><div style="text-align: justify;"><span style="font-family: Times; font-size: medium;"><br /></span></div><div style="text-align: justify;"><span style="font-family: Times; font-size: medium;">$\Large R_{P} = R_{S}\sqrt\frac{T_{depth}}{1,049}$</span></div><div style="text-align: justify;"><br /></div>

```python
def raio(trans_depth, raio_estrela):
    raio_planeta = ((trans_depth*raio_estrela**2)/1.049)**(1/2)
    return raio_planeta
```

<div style="text-align: justify;"><b><span style="font-family: Times; font-size: large;">Probabilidade de trnsito</span></b></div><div style="text-align: justify;"><span style="font-family: Times; font-size: medium;">Usando as coordenadas polares, x pode ser escrito da seguinte forma</span></div><div style="text-align: justify;"><span style="font-family: Times; font-size: medium;"><br /></span></div><div style="text-align: justify;"><span style="font-family: Times; font-size: medium;">$\Large x = rcos(i)$<br /><br /></span></div><div style="text-align: justify;"><span style="font-family: Times; font-size: medium;">Onde r = a. A condio necessria para que o trnsito ocorra <br /><br /></span></div><div style="text-align: justify;"><span style="font-family: Times; font-size: medium;">$\Large acos(i) < R_{S} + R_{P}$<br /><br /></span></div><div style="text-align: justify;"><span style="font-family: Times; font-size: medium;">A partir dessa ideia, o trnsito pode ocorrer at o limite do tamanho estelar, ento, a probabilidade que ele ocorra  definida por&nbsp;<br /><br /></span></div><div style="text-align: justify;"><span style="font-family: Times; font-size: medium;">$\Large P_{tr} = \frac {\int_{0}^{mx cos(i)}d(cos(i))}{\int_{0}^{1}d(cos(i))}$<br /></span></div><div style="text-align: justify;"><span style="font-family: Times; font-size: medium;"><br /></span></div><div style="text-align: justify;"><span style="font-family: Times; font-size: medium;">Como o mximo valor que cos(i) pode assumir </span></div><div style="text-align: justify;"><span style="font-family: Times; font-size: medium;"><br /></span></div><div style="text-align: justify;"><span style="font-family: Times; font-size: medium;">$\Large cos(i) = \frac {R_{S}+R_{P}}{a}$</span></div><div style="text-align: justify;"><span style="font-family: Times; font-size: medium;"><br /></span></div><div style="text-align: justify;"><span style="font-family: Times; font-size: medium;">Tem- se que</span></div><div style="text-align: justify;"><span style="font-family: Times; font-size: medium;"><br /></span></div><div style="text-align: justify;"><span style="font-family: Times; font-size: medium;">$\Large P_{tr} = \frac {R_{S}+R_{P}}{a}\frac{\int_{0}^{1}d(cos(i))}{\int_{0}^{1}d(cos(i))}$</span></div><div style="text-align: justify;"><span style="font-family: Times; font-size: medium;"><br /></span></div><div style="text-align: justify;"><span style="font-family: Times; font-size: medium;">Sabendo que a funo cos tem seu mximo em 1, a equao anterior pode ser escrita da seguinte forma</span></div><div style="text-align: justify;"><span style="font-family: Times; font-size: medium;"><br /></span></div><div style="text-align: justify;"><span style="font-family: Times; font-size: medium;">$\Large P_{tr} = \frac {R_{S}+R_{P}}{a}$</span></div><div style="text-align: justify;"><span style="font-family: Times; font-size: medium;"><br /></span></div><div style="text-align: justify;"><span style="font-family: Times; font-size: medium;">E assumindo que Rs &gt;&gt; Rp</span></div><div style="text-align: justify;"><span style="font-family: Times; font-size: medium;"><br /></span></div><div style="text-align: justify;"><span style="font-family: Times; font-size: medium;">$\Large P_{tr} = \frac {R_{S}}{a}$</span></div><div style="text-align: justify;"><span style="font-family: Times; font-size: medium;"><br /></span></div><div style="text-align: justify;"><span style="font-family: Times; font-size: medium;">Mas para esse resultado, no  levada em considerao a excentricidade, pode-se considerar que</span></div><div style="text-align: justify;"><span style="font-family: Times; font-size: medium;"><br /></span></div><div style="text-align: justify;"><span style="font-family: Times; font-size: medium;">$\Large r = \frac{a(1-e^{2})}{1+ecos()}$</span></div><div style="text-align: justify;"><span style="font-family: Times; font-size: medium;"><br /></span></div><div style="text-align: justify;"><span style="font-family: Times; font-size: medium;">Ento</span></div><div style="text-align: justify;"><span style="font-family: Times; font-size: medium;"><br /></span></div><div style="text-align: justify;"><span style="font-family: Times; font-size: medium;">$\Large P_{tr} = \frac {R_{S}+R_{P}(1+ecos())}{a(1-e^{2})}$</span></div><div style="text-align: justify;"><span style="font-family: Times; font-size: medium;"><br /></span></div><div style="text-align: justify;"><span style="font-family: Times; font-size: medium;">E considerando que o cos(<b style="background-color: white; text-align: left;"></b>)  aproximadamente igual a 0, a equao para a probabilidade de trnsito </span></div><div style="text-align: justify;"><span style="font-family: Times; font-size: medium;"><br /></span></div><div style="text-align: justify;"><span style="font-family: Times; font-size: medium;">$\Large P_{tr} = \frac {R_{S}+R_{P}}{a(1-e^{2})}$</span></div>

```python
def probabilidade_transito(raio_estrela, semieixo, excentricidade):
    prob = (raio_estrela*raio_sol)/(semieixo*UA*(1-excentricidade)**2)
    return prob*100
```

<span style="font-family: Times; font-size: large;"><b>Massa do planeta<br /></b></span><div><span style="font-family: Times; font-size: medium;">Imaginando que o planeta e a estrela possuem o mesmo perodo orbital, logo, eles teriam o mesmo momento, ento, o seguinte  vlido</span></div><div><span style="font-family: Times; font-size: medium;"><br /></span></div><div><span style="font-family: Times; font-size: medium;">$\Large M_{S}v_{S} = M_{P}v_{P}$</span></div><div><span style="font-family: Times; font-size: medium;"><br /></span></div><div><span style="font-family: Times; font-size: medium;">Determinando a massa e a velocidade da estrela, resta apenas trabalhar na velocidade do planeta, para determinar sua massa, e utilizando a equao anterior, para determinar o semieixo maior, tal que</span></div><div><span style="font-family: Times; font-size: medium;"><br /></span></div><div><span style="font-family: Times; font-size: medium;">$\Large v_{P} = \frac {2a}{P}$</span></div><div><span style="font-family: Times; font-size: medium;"><br /></span></div><div><span style="font-family: Times; font-size: medium;">Acima, a velocidade do planeta,  dada pela distncia que ele percorre, circunferncia da orbita, sobre o tempo em que ele faz esse caminho, o perodo, que  observvel, substituindo a velocidade, na primeira equao, e usando a velocidade da estrela como velocidade radial,  possvel encontrar que</span></div><div><span style="font-family: Times; font-size: medium;"><br /></span></div><div><span style="font-family: Times; font-size: medium;">$\Large M_{P}= \frac {M_{S}KP}{2a}$</span></div>

```python
def massa(massa_estrela, amp_velocidade, periodo, semieixo):
    massa_planeta = (massa_estrela*amp_velocidade*periodo)/(2*sci.pi*semieixo)
    return massa_planeta
```

<span style="font-family: Times;"><b style="font-size: x-large;">Tempo de trnsito<br /></b><span style="font-family: Times; font-size: medium;">Tempo que o planeta leva para atravessar completamente sua estrela.</span></span><div><span style="font-family: Times; Ento, por trigonometria, tem-se</span></span></span></div><div><span style="font-size: medium;"><span style="background-color: white;"><span style="font-family: Times;"><br /></span></span></span></div><div><span style="background-color: white;"><span style="font-family: Times; font-size: medium;">$\Large sen\frac{}{2} = \frac {I}{sen(i)}$</span></span></div><div><span style="background-color: white;"><span style="font-family: Times; font-size: medium;"><br /></span></span></div><div><span style="background-color: white;"><span style="font-family: Times; font-size: medium;">Substituindo 1 pela equao anterior</span></span></div><div><span style="background-color: white;"><span style="font-family: Times; font-size: medium;"><br /></span></span></div><div><span style="background-color: white;"><span style="font-family: Times; font-size: medium;">$\Large sen\frac{}{2} = \frac {((R_{P}+R_{S})^2 - (acos(i))^{2})^{1/2}}{asen(i)}$</span></span></div><div><span style="background-color: white;"><span style="font-family: Times; font-size: medium;"><br /></span></span></div><div><span style="background-color: white;"><span style="font-family: Times; font-size: medium;">E reorganizando a equao, obtem-se</span></span></div><div><span style="background-color: white;"><span style="font-family: Times; font-size: medium;"><br /></span></span></div><div><span style="background-color: white;"><span style="font-family: Times; font-size: medium;">$\Large sen\frac{}{2} = \frac {R_{S}}{a}\frac {((1+ \frac {R_{P}}{R_{S}})^2 - (\frac {acos(i)}{R_{S}})^{2})^{1/2}}{1-cos(i)^2}$</span></span></div><div><span style="font-size: medium;"><span style="background-color: white;"><span style="font-family: Times;"><br /></span></span></span></div><div><span style="font-size: medium;"><span style="background-color: white;"><span style="font-family: Times;">Como o&nbsp;</span></span><span style="font-family: Times;">ngulo&nbsp;</span><span style="background-color: white;"><span style="font-family: Times;">  relacionado com o comprimento da circunferncia pelo perodo, onde</span></span></span></div><div><span style="font-size: medium;"><span style="background-color: white;"><span style="font-family: Times;"><br /></span></span></span></div><div><span style="background-color: white;"><span style="font-family: Times; font-size: medium;">$\Large  = \frac {2t_{r}}{P}$</span></span></div><div><span style="background-color: white;"><span style="font-family: Times; font-size: medium;"><br /></span></span></div><div><span style="background-color: white;"><span style="font-family: Times; font-size: medium;">Ento</span></span></div><div><span style="background-color: white;"><span style="font-family: Times; font-size: medium;">$\Large sen\frac {t_{r}}{P}$</span></span></div><div><span style="background-color: white;"><span style="font-family: Times; font-size: medium;"><br /></span></span></div><div><span style="background-color: white;"><span style="font-family: Times; font-size: medium;">E por consequncia, isolando Tr  dado por</span></span></div><div><span style="background-color: white;"><span style="font-family: Times; font-size: medium;"><br /></span></span></div><div><span style="background-color: white;"><span style="font-family: Times; font-size: medium;">$\Large T_{t} = \frac{P}{}arcsen(\frac{R_{S}}{a}[\frac{(1-\frac {R_{P}}{R_{S}})^{2} - (\frac{a}{R_{S}}cos(i)^{2})^2}{1-cos(i)^{2}}]^{1/2})$</span></span></div>

```python
def tempo_transito(periodo, raio_estrela, semieixo, raio_planeta, inclinacao):
    parte1 = ((semieixo/raio_estrela)*(math.cos(math.radians(inclinacao))**2))**2
    parte2 = 1-(math.cos(math.radians(inclinacao))**2)
    parte3 = (1-raio_planeta/raio_estrela)**2
    final = (periodo/math.pi)*math.asin((raio_estrela/semieixo)*(((parte3-parte1)/parte2)**(1/2)))
    return final
```

<p style="text-align: justify;"><span style="font-family: Times;"><span style="font-size: large;"><b>Semieixo maior do baricentro do sistema<br /></b></span><span style="font-size: medium;">Um baricentro, ou centroide, para a geometria,  o ponto em que todas as medianas de um polgono&nbsp;se encontram, j do ponto de vista astrofsico, ele seria o centro da gravidade, ou mais conhecido como centro de massa, ponto onde as massas esto em equilbrio, num sistema planetrio,  o local em torno do qual os corpos do sistema orbitam. Fazendo a aproximao das rbitas como esfrica, usa-se essa aproximao para ser determinado.</span></span></p><p style="text-align: justify;"><span style="font-family: Times;"><span style="font-size: medium;">Fazendo certas consideraes, buscando facilitar o problema, e usando coordenadas cartesianas a posio da estrela pode ser dada como (0,0), a posio do exoplaneta como (x, 0). E utilizando as noes bsicas do centro de massa, chegasse a seguinte relao para encontrar o semieixo do baricentro.</span></span></p><p style="text-align: justify;"><span><span style="font-size: medium;">$\large B = \frac {P(1-e^{2})^{\frac{1}{2}}K}{2sen(i)}$</span><br /><br /></span></p>

```python
def baricentro(periodo, excentricidade, velocidade_radial, inclinacao):
    bar_sma = (periodo*((1-(excentricidade**2))**(1/2))*velocidade_radial)/(2*sci.pi*math.sin(math.radians(inclinacao)))
    return bar_sma
```
